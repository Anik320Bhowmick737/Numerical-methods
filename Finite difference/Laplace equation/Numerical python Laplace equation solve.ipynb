```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid=10
phi=np.zeros((grid+1,grid+1))
```

```python
phi
```

```python
def BC(x,y):
    return x**2-y**2
```

```python
for i in range(len(phi)):
    phi[0,i]=BC(0,i/10)
    phi[10,i]=BC(1,i/10)
    phi[i,0]=BC(i/10,0)
    phi[i,10]=BC(i/10,1)
```

```python
phi
```

```python
iteration=0
while iteration<10000:
    for i in range(1,len(phi)-1):
        for j in range(1,len(phi)-1):
            phi[i,j]=0.25*(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])  
    iteration=iteration+1        
phi=np.transpose(phi)
phi
```

```python
x,y=np.arange(11)/10,np.arange(11)/10
X,Y=np.meshgrid(x,y)
X**2-Y**2
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, X**2-Y**2, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, phi, cmap='magma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
