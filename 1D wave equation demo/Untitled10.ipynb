```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=np.zeros((1001,1001))
```

```python
x=np.linspace(0,1,1001)
t=np.linspace(0,10,1001)
```

a=0.05 m/s  
$\Delta t = $0.01 s  
CFL=$\frac{0.05\times 0.01}{0.001}=0.5$

```python
u[0,:]=np.sin(np.pi*x)#Initial condition specified
```

```python
u
```

```python
#Visualisation
#@t=0
plt.plot(x,u[0,:],label='@t=0')
plt.legend()
plt.show()
```

```python
def FTBS(U,CFL):
    for i in range(len(U[:,0])-1):
        for j in range(len(U[0,:])-1):
            U[i+1,j+1]=U[i,j+1]-CFL*(U[i,j+1]-U[i,j])
    return U        
```

```python
FTBS(u,0.5)
```

```python
a=np.arange(0,1001,100)
for i in a:
    plt.plot(x,u[i,:],label="@ t={}".format(i/100))
    plt.grid()
    plt.legend()
    plt.show()
```
